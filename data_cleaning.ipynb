```python
import pandas as pd
```

```python
# Cleaning "mvps" Data Set
mvp = pd.read_csv("mvps.csv")
```

```python
mvp
```

```python
# Fix Special Character Naming Issue
mvp.replace('Luka DonÄiÄ', 'Luka Dončić', inplace=True)
mvp.replace('Nikola JokiÄ','Nikola Jokić', inplace=True)
mvp.replace('Manu GinÃ³bili','Manu Ginóbili', inplace=True)
mvp.replace('Peja StojakoviÄ','Peja Stojaković', inplace=True)
mvp.replace('Goran DragiÄ','Goran Dragić', inplace=True)
# Remove Unnecessary Columns
mvps = mvp[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.tail(10)
```

```python
# Cleaning The PLayer Data
players = pd.read_csv("players.csv")
players.head()
```

```python
# Some players have an asterisk, so we must remove it
players["Player"].head(24)

# Merge player Dataframe with mvp dataframe
```

```python
# Lets remove the asterisk
players["Player"] = players["Player"].str.replace("*", "", regex=False)
players["Player"].head(24)
# Now its gone!
```

```python
players.head(20)
```

```python
# Create Function named "Single Row" to iterate through each group so that it only has one row
# Replace "TOT" with the last team he played for
players[players["Player"] == "Greg Anderson"]
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
# Now we merge the two dataframes (Outer Merge keeps data if not found in the other set)
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
combined.head(5)
```

```python
combined[combined["Pts Won"] > 0]
```

```python
# Fix NaN Values 
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
combined
```

```python
# Now Lets Clean the teams csv
teams = pd.read_csv("teams.csv")
```

```python
teams.head(30)
```

```python
# Filter out rows from the 'teams' DataFrame where the value in the 'W' column contains 'Division'
teams = teams[~teams["W"].str.contains("Division")].copy()
# Get rid off Asteriks 
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(5)
```

```python
# The full name is listed in the Team DF while....
sorted(teams["Team"].unique())
```

```python
# The combined df has abbreviations for team names
sorted(combined["Tm"].unique())
```

```python
# Map Team name with abbreviation using 
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        parts = line.replace("\n", "").split(",")
        if len(parts) == 2:
            abbrev, name = parts
            nicknames[abbrev] = name
        else:
            print(f"Ignoring line: {line}")

```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
# Now we can merge
train = combined.merge(teams, how="outer",on=["Team", "Year"])

train
```

```python
# Delete unwanted column
del train["Unnamed: 0"]
```

```python
# Check Data Type of Column
train.dtypes
```

```python
# Convert object that should be floats
train = train.apply(pd.to_numeric, errors='ignore')
train.dtypes
```

```python
# GB should be an integer, so lets change the '-' to a 0
train["GB"].unique()
train["GB"] = pd.to_numeric(train["GB"].str.replace("—","0"))
```

```python
train.dtypes
```

```python
train.to_csv("player_mvp_stats.csv")
```

```python
# Who Scored the most points?
highest_scoring = train[train["G"] > 70].sort_values("PTS", ascending=False).head(10)
highest_scoring.plot.bar("Player", "PTS")
```
